```python
import pandas as pd
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('data/data.csv')
```

```python
tuition = np.array(data['Tuition and fees'])
tuition = tuition[~np.isnan(tuition)]

sns_plot = sns.distplot(tuition, kde=False, axlabel='Cost of Tuition and Fees (USD)')
fig = sns_plot.get_figure()
fig.savefig("visualizations/tuition_hist.png")
```

```python
sns_plot = sns.scatterplot(x='Tuition and fees', y='Admissions yield', data=data)
fig = sns_plot.get_figure()
fig.savefig("visualizations/tuition_admissions.png")
```

```python
sns_plot = sns.scatterplot(x='Tuition and fees', y='Graduation rate', data=data)
fig = sns_plot.get_figure()
fig.savefig("visualizations/tuition_grad_rate.png")
```

```python
sns_plot = sns.scatterplot(x='Tuition and fees', y='Average professor salary', data=data)
fig = sns_plot.get_figure()
plt.tight_layout()
fig.savefig("visualizations/tuition_salary.png")
```
